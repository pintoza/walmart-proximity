```python
data_encoded = pd.get_dummies(data, columns=['ruca'])
data = data_encoded
```

```python
# Including great circle distance in the model
X = data[['great_circle_distance', 'ruca_Rural', 'ruca_Suburban', 'ruca_Urban']]
y = data['driving_distance']

# Create the regression model
model = LinearRegression()
model.fit(X, y)

# Predict and calculate R-squared for the combined model
predictions = model.predict(X)
r_squared = r2_score(y, predictions)
print("R-squared with Great Circle Distance and RUCA classifications:", r_squared)
```

```python
# Coefficients
coefficients = pd.DataFrame(model.coef_, X.columns, columns=['Coefficient'])
print(coefficients)

# Intercept
print(f"Intercept: {model.intercept_}")
```

```python
# If you have a holdout test set or can perform cross-validation, do so here
from sklearn.model_selection import cross_val_score
cross_val_scores = cross_val_score(model, X, y, cv=5, scoring='r2')
print("Cross-validated R-squared scores:", cross_val_scores)
```

```python
from sklearn.linear_model import LinearRegression

X = data[['great_circle_distance', 'ruca_Rural', 'ruca_Suburban', 'ruca_Urban']]  # or any other predictors you want to include
y = data['driving_distance']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, y_pred)
plt.xlabel('Actual Driving Distance')
plt.ylabel('Predicted Driving Distance')
plt.title('Actual vs Predicted Driving Distance')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)  # Diagonal line
plt.show()
```

```python
residuals = y_test - y_pred

plt.scatter(y_pred, residuals)
plt.xlabel('Predicted Driving Distance')
plt.ylabel('Residuals')
plt.title('Residuals vs Predicted Driving Distance')
plt.axhline(y=0, color='red', linestyle='--')
plt.show()
```

```python
from sklearn.model_selection import cross_val_score

cross_val_scores = cross_val_score(model, X, y, cv=5, scoring='r2')
print("Cross-validated R-squared scores:", cross_val_scores)
```

```python
# Now we predict the distances using the trained model.
predicted_distances = model.predict(X)

# Use the conservative threshold to generate binary labels.
CONSERVATIVE_THRESHOLD = 8  # Adjust as needed.
binary_labels = np.where(predicted_distances <= (CONSERVATIVE_THRESHOLD * 1.60934), 1, 0)

# Add these binary labels to your dataframe.
data['within_10_miles'] = binary_labels
```

```python
import matplotlib.pyplot as plt

# Assuming 'actual_distances' is a list or array containing the actual driving distances.
# And 'binary_labels' contains the binary classification results as described above.
actual_distances = data['driving_distance']
binary_labels = data['within_10_miles']

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(actual_distances, binary_labels, alpha=0.5)

# Add title and labels
plt.title('Binary Classifier vs Actual Driving Distance')
plt.xlabel('Actual Driving Distance (miles)')
plt.ylabel('Binary Classification (Within 10 miles)')

# Define the x-axis limits
plt.xlim(0, max(actual_distances))

# Define the y-axis limits to be just enough to see the binary nature
plt.ylim(-0.1, 1.1)

# Add a horizontal line at y=0.5 to delineate the classification boundary
plt.axhline(y=0.5, color='r', linestyle='--')

# Show the plot
plt.show()
```
